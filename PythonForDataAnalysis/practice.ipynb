# 계단 오르기
동전을 던져서 앞면(1)이 나오면 한칸을 올라가고  
뒷면(0)이 나오면 한칸을 내려간다.  
동전을 500번 던졌을 때 최종 위치를 구하고,  
10번째 계단에 최초로 도착했을때의 동전 던진 횟수를 구해라. (도착 못 했다면 0 출력)

```python
import numpy as np
```

```python
rng = np.random.default_rng(seed=12345)
head_tail_arr = rng.integers(0, 2, 500)
# print(head_tail_arr)
```

```python
up_down_arr = np.where(head_tail_arr, 1, -1)
# print(up_down_arr)
```

```python
current_stair_arr = up_down_arr.cumsum()
# print(current_stair_arr)
```

```python
print(f'최종 위치    : {current_stair_arr[-1]}')
print(f'최초 10 도착 : {(current_stair_arr == 10).argmax()}')
```

## 100명이 동시에 할 때,  
## 10 위치에 도착했던 인원 수와 10 위치에 도착한 사람들의 최종 위치를 출력해라.

```python
import numpy as np
rng = np.random.default_rng(seed = 12345)
head_tail_arr = rng.integers(0, 2, (100, 500))
print(head_tail_arr)
```

```python
up_down_arr = np.where(head_tail_arr == 1, 1, -1)
print(up_down_arr)
```

```python
current_stair_arr = up_down_arr.cumsum(1)
print(current_stair_arr)
```

```python
for i in range(100) :
    print(f'{i + 1} 번째 사람')
    print(f'최종 위치    : {current_stair_arr[i][-1]}')
    print(f'최초 10 도착 : {(current_stair_arr[i] == 10).argmax()}')
    print()
```

```python
pass_10_person_arr = (current_stair_arr == 10).any(axis=1)

print(f'10 도착 인원 : {pass_10_person_arr.sum()} 명')

# final_stair_arr =  []
# for i in range(pass_10_person_arr.size) :
#     if pass_10_person_arr[i] :
#         final_stair_arr.append(current_stair_arr[i][-1])
# print(final_stair_arr)

# 리스트 컴프리헨션을 이용한 방법
final_stair_arr = [current_stair_arr[i][-1] for i in range(pass_10_person_arr.size) if pass_10_person_arr[i]]

print(final_stair_arr)
```
