### 57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라  
Unknown : N  
Less than $40K : a  
$40K - $60K : b  
$60K - $80K : c  
$80K - $120K : d  
$120K +’ : e

````python
dic = {
    'Unknown'           : 'N',
    'Less than $40K'    : 'a',
    '$40K - $60K'       : 'b',
    '$60K - $80K'       : 'c',
    '$80K - $120K'      : 'd',
    '$120K +'           : 'e'
}

df['newIncome'] = df['Income_Category'].map(lambda x : dic[x])
df['newIncome'].head()
````

### 58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라  
Unknown : N  
Less than $40K : a  
$40K - $60K : b  
$60K - $80K : c  
$80K - $120K : d  
$120K +’ : e

````python
def changeCategory(x):
    if x == 'Unknown' :
        return 'N'
    elif x == 'Less than $40K' :
        return 'a'
    elif x == '$40K - $60K' :
        return 'b'
    elif x == '$60K - $80K' :
        return 'c'
    elif x == '$80K - $120K' :
        return 'd'
    else :
        return 'e'

df['newIncome'] = df['Income_Category'].apply(changeCategory)
df['newIncome'].head()
````

### 63. Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하라. 각 value의 빈도를 출력하라

````python
df['Gender'].head()
````

````python
#df['Gender'] = df.apply(lambda x : 'male' if x['Gender'] == 'M' else 'female', axis=1)
#df['Gender'] = df['Gender'].apply(lambda x : 'male' if x == 'M' elif x == 'F': 'female')
df['Gender'].value_counts()
````

```
lambda에서는 elif를 쓸 수 없다.  
위 명령어를 2번 실행하면 male이 전부 else로 들어가서 전부 female이 되어 버린다.  
조건이 간단할 때만 lambda를 사용하고, 다른 때는 def를 이용해 함수를 만들어서 사용하자.
```

````python
def changeGender(x):
    if x == 'M':
        return 'male'
    elif x == 'F':
        return 'female'
    
df['Gender'] = df['Gender'].apply(changeGender)
df['Gender'].value_counts()
````

### 76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다  
서울시 미세먼지 데이터 : https://www.airkorea.or.kr/web/realSearch?pMENU_NO=97  
DataUrl = ‘https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv’

````python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df.head()
````

````python
def change_date(x):
    import datetime
    date = x.split(':')[0]
    hour = x.split(':')[1]
    if hour == '24':
        hour = '00:00:00'
    else :
        hour += ':00:00'
    final_date = date + " " + hour
    return pd.to_datetime(final_date)

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head()
````

columns: 새로운 데이터프레임의 열 이름으로 사용할 열의 이름
